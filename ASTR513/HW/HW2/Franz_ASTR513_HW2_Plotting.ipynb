```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson, binom, expon, uniform, moment

import seaborn as sb
sb.set(context='paper', style='white', palette='Set1')
plt.rcParams["font.family"] = "serif"
plt.rcParams["figure.dpi"] = 200
plt.rcParams["axes.grid.which"] = 'both'
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['xtick.major.size'] = 3
plt.rcParams['ytick.major.size'] = 3

seed = 42
```

## Generate Plots of each distribution
### Normal

```python
fig, ax = plt.subplots()

# true distribution
x = np.linspace(-5,5,200)
y = norm.pdf(x)
ax.plot(x, y, 'k', label='True Normal Distribution')

n1000 = norm.rvs(size=1000, random_state=seed)
n100000 = norm.rvs(size=100000, random_state=seed)

# samples
ax.hist(n1000, density=True, bins=100, histtype='stepfilled', alpha=0.4, label='N=1e4 Samples');
ax.hist(n100000, density=True, bins=100, histtype='stepfilled', alpha=0.4, label='N=1e6 Samples');

ax.set_ylabel("Probability Density")
ax.set_xlabel("X")

ax.legend();

fig.savefig('normal.jpeg', bbox_inches='tight', transparent=False)
```

```python
for val in [n1000, n100000]:
    for i in range(1,5):
        print(moment(val, i))
    print()
```

### Poisson

```python
fig, ax = plt.subplots()

mu = 50

# true distribution (just generate a lot of samples)
x = np.arange(0,mu*3)
ax.plot(x, poisson.pmf(x, mu), label='True Poisson Distribution', color='k')
# samples
ax.hist(poisson.rvs(mu=mu, size=1000, random_state=seed), density=True, bins=mu, histtype='stepfilled', alpha=0.4, label='N=1e4 Samples');
ax.hist(poisson.rvs(mu=mu, size=100000, random_state=seed), density=True, bins=mu, histtype='stepfilled', alpha=0.4, label='N=1e6 Samples');

ax.set_ylabel("Probability Density")
ax.set_xlabel("X")

ax.legend();
fig.savefig('poisson.jpeg', bbox_inches='tight', transparent=False)
```

### Binomial

```python
fig, ax = plt.subplots()

n = 100
p = 0.5
b = binom(n,p)

# true distribution (just generate a lot of samples)
x = np.arange(0,n,1)
ax.plot(x, b.pmf(x), label='True Binomial Distribution', color='k')
# samples
ax.hist(b.rvs(size=1000, random_state=seed), density=True, bins=n//2, histtype='stepfilled', alpha=0.4, label='N=1e4 Samples');
ax.hist(b.rvs(size=100000, random_state=seed), density=True, bins=n//2, histtype='stepfilled', alpha=0.4, label='N=1e6 Samples');

ax.set_ylabel("Probability Density")
ax.set_xlabel("X")

ax.legend();
fig.savefig('binomial.jpeg', bbox_inches='tight', transparent=False)
```

### Exponential

```python
fig, ax = plt.subplots()

# true distribution
x = np.linspace(0,10,200)
y = expon.pdf(x)
ax.plot(x, y, 'k', label='True Exponential Distribution')

# samples
e1000 = expon.rvs(size=1000, random_state=seed)
e100000 = expon.rvs(size=100000, random_state=seed)
ax.hist(e1000, density=True, bins=100, histtype='stepfilled', alpha=0.4, label='N=1e4 Samples');
ax.hist(e100000, density=True, bins=100, histtype='stepfilled', alpha=0.4, label='N=1e6 Samples');

ax.set_ylabel("Probability Density")
ax.set_xlabel("X")

ax.legend();
fig.savefig('exponential.jpeg', bbox_inches='tight', transparent=False)
```

```python
for val in [e1000, e100000]:
    for i in range(1,5):
        print(moment(val, i))
    print()
```

### Uniform

```python
fig, ax = plt.subplots()

# true distribution
x = np.linspace(0,1,200)
y = uniform.pdf(x)
ax.plot(x, y, 'k', label='True Uniform Distribution')

# samples
ax.hist(uniform.rvs(size=1000, random_state=seed), density=True, bins=100, histtype='stepfilled', alpha=0.4, label='N=1e4 Samples');
ax.hist(uniform.rvs(size=100000, random_state=seed), density=True, bins=100, histtype='stepfilled', alpha=0.4, label='N=1e6 Samples');

ax.set_ylabel("Probability Density")
ax.set_xlabel("X")

ax.legend();
fig.savefig('uniform.jpeg', bbox_inches='tight', transparent=False)
```
